```python
import scrapy
import subprocess
import os
```

```python
import json
def read_output(json_path, get_collaborator_links=False):
    l = json.loads(open(json_path).read())
    out = {}
    for entry in l:
        if get_collaborator_links:
            for indx, name in enumerate(entry['Collaborators']):
                out[name] = entry['CollaboratorsLink'][indx]
    
    return out
        
```

```python

cmd_template = "scrapy runspider --nolog scrape.py -a start_url_file='./out/$infile' -O ./out/$outfile"

depth = 3
for i in range(depth):
    # run the iteration
    print('RUNNING ITERATION #%s' %i)
    print('RUNNING %s SCHOLARS:' %len(open('./out/%s.txt' %i).readlines()))
    cmd = cmd_template.replace('$infile', '%s.txt' %i).replace('$outfile', '%s.json' %i)
    print('COMMAND:')
    print('\t' + cmd)
    # p = subprocess.Popen(cmd.split(), stdout=subprocess.PIPE)
    # out, err = p.communicate()
    os.system(cmd)
    
    # read and write new url file
    print('WRITING URLS EXTRACTED FROM THE SCRAPE:')
    out = read_output('./out/%s.json' %i, get_collaborator_links=True)
    base_url = 'https://scholar.google.com'
    s = ''
    for k, v in out.items():
        s += base_url + v + ' # ' + k + '\n'
    with open('./out/%s.txt' %(i+1), 'w') as f:
        f.write(s)
```

```python
# aggregate all the produced jsons into one file
files = [q for q in os.listdir('./out/') if '.json' in q]
n = []
print(files)
for i in files[1:-1]:
    n.extend(json.loads(open('./out/'+i).read()))

# filter duplicate entries
outn = []
done_names = []
for i in n:
    if i['Name'] not in done_names:
        outn.append(i)
    done_names.append(i['Name'])

with open('./out/agg.json', 'w') as f:
    f.write(json.dumps(n))
```
